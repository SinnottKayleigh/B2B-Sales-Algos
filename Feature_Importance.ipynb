Feature importance increases the accuracy of prediction for an ML model by selecting the most important variables and subsequently eliminating the redunandant ones.
- Reduces overfitting
- Improves model generalisation to data

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

class FeatureImportanceAnalyzer:
    def __init__(self, data):
        self.data = data
        self.feature_importance_scores = None
        self.model = None

    def analyze_feature_importance(self):
        features = [
            'company_size',
            'international_revenue_percentage',
            'number_of_countries',
            'industry_sector',
            'current_fx_provider',
            'annual_fx_volume',
            'response_rate',
            'engagement_score',
            'website_visits',
            'content_downloads'
        ]

        model = RandomForestClassifier(n_estimators=100)
        importance_scores = model.feature_importances_

        return pd.DataFrame({
            'feature': features,
            'importance': importance_scores
        }).sort_values('importance', ascending=False)

    def visualize_importance(self):
        plt.figure(figsize=(12, 6))
        sns.barplot(
            data=self.feature_importance_scores,
            x='importance',
            y='feature'
        )
        plt.title('Feature Importance in Lead Conversion')
        plt.show()

    def get_key_insights(self):
        """Extract actionable insights from feature importance"""
        insights = {
            'high_impact_features': self.feature_importance_scores.head(3)['feature'].tolist(),
            'low_impact_features': self.feature_importance_scores.tail(3)['feature'].tolist(),
            'recommendations': []
        }
        return insights
```

```python
def analyze_feature_roi(feature_importance, cost_data, conversion_data):
    roi_analysis = {}

    for feature, importance in feature_importance.items():
        cost = cost_data.get(feature, 0)
        impact = importance * conversion_data['conversion_rate']
        roi = (impact - cost) / cost if cost > 0 else 0

        roi_analysis[feature] = {
            'importance_score': importance,
            'cost': cost,
            'impact': impact,
            'roi': roi,
            'recommendation': get_roi_recommendation(roi)
        }

    return roi_analysis
```

Feature importance change algorithm, to track the changes in feature importance within data overtime.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from datetime import datetime, timedelta

class FXFeatureImportanceTracker:
    def __init__(self):
        self.historical_importance = pd.DataFrame()
        self.current_importance = None
        self.change_metrics = None

    def calculate_feature_importance(self, data, target, date):
        """Calculate feature importance for a specific time period"""
        model = RandomForestClassifier(n_estimators=100, random_state=42)
        model.fit(data, target)

        importance_df = pd.DataFrame({
            'feature': data.columns,
            'importance': model.feature_importances_,
            'date': date
        })

        return importance_df

    def track_importance_changes(self, new_importance_data):
        """Track changes in feature importance over time"""
        self.historical_importance = pd.concat([
            self.historical_importance,
            new_importance_data
        ])

        self.change_metrics = self.calculate_change_metrics()

    def calculate_change_metrics(self):
        """Calculate various change metrics for features"""
        latest_date = self.historical_importance['date'].max()
        previous_date = self.historical_importance['date'].unique()[-2]

        latest_data = self.historical_importance[
            self.historical_importance['date'] == latest_date
        ]
        previous_data = self.historical_importance[
            self.historical_importance['date'] == previous_date
        ]

        change_metrics = pd.DataFrame({
            'feature': latest_data['feature'],
            'current_importance': latest_data['importance'],
            'previous_importance': previous_data['importance'],
            'absolute_change': latest_data['importance'] - previous_data['importance'],
            'percentage_change': ((latest_data['importance'] - previous_data['importance'])
                                / previous_data['importance'] * 100)
        })

        return change_metrics

    def generate_trend_analysis(self):
        """Analyze trends in feature importance"""
        trends = {}

        for feature in self.historical_importance['feature'].unique():
            feature_data = self.historical_importance[
                self.historical_importance['feature'] == feature
            ]

            trend = np.polyfit(
                range(len(feature_data)),
                feature_data['importance'],
                deg=1
            )[0]

            volatility = feature_data['importance'].std()

            trends[feature] = {
                'trend_direction': 'Increasing' if trend > 0 else 'Decreasing',
                'trend_strength': abs(trend),
                'volatility': volatility,
                'stability_score': 1 / (1 + volatility)  # Higher score = more stable
            }

        return trends

    def generate_alerts(self):
        """Generate alerts for significant changes"""
        alerts = []

        for _, row in self.change_metrics.iterrows():
            if abs(row['percentage_change']) > 20:  # Significant change threshold
                alerts.append({
                    'feature': row['feature'],
                    'alert_type': 'Significant Change',
                    'details': f"{row['percentage_change']:.1f}% change in importance",
                    'priority': 'High' if abs(row['percentage_change']) > 50 else 'Medium'
                })

        return alerts

    def generate_report(self):
        """Generate comprehensive report"""
        report = {
            'summary_metrics': {
                'most_important_features': self.current_importance.nlargest(3, 'importance'),
                'biggest_changes': self.change_metrics.nlargest(3, 'absolute_change'),
                'most_stable_features': self.generate_trend_analysis()
            },
            'trend_analysis': self.generate_trend_analysis(),
            'alerts': self.generate_alerts(),
            'recommendations': self.generate_recommendations()
        }

        return report

    def generate_recommendations(self):
        """Generate recommendations based on feature importance changes"""
        recommendations = []

        for _, row in self.change_metrics.iterrows():
            if row['percentage_change'] > 20:
                recommendations.append({
                    'feature': row['feature'],
                    'action': 'Increase focus',
                    'reasoning': 'Significant increase in importance',
                    'suggested_steps': [
                        'Update lead scoring weights',
                        'Adjust sales qualification criteria',
                        'Modify data collection priorities'
                    ]
                })
            elif row['percentage_change'] < -20:
                recommendations.append({
                    'feature': row['feature'],
                    'action': 'Review and assess',
                    'reasoning': 'Significant decrease in importance',
                    'suggested_steps': [
                        'Analyze cause of decrease',
                        'Consider reducing resource allocation',
                        'Update sales playbook'
                    ]
                })

        return recommendations

tracker = FXFeatureImportanceTracker()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

class EnhancedFXSalesDashboard:
    def __init__(self):
        self.initialize_data()
        self.train_predictive_models()

    def initialize_data(self):
        dates = pd.date_range(start='2024-01-01', end=datetime.now(), freq='D')
        n_dates = len(dates)

        self.data = pd.DataFrame({
            'date': dates,
            'leads_generated': np.random.normal(100, 20, n_dates),
            'conversion_rate': np.random.normal(0.15, 0.03, n_dates),
            'fx_volume': np.random.normal(1000000, 200000, n_dates),
            'revenue': np.random.normal(50000, 10000, n_dates),
            'spot_volume': np.random.normal(700000, 150000, n_dates),
            'forward_volume': np.random.normal(300000, 80000, n_dates),
            'option_volume': np.random.normal(100000, 30000, n_dates),
            'avg_trade_size': np.random.normal(50000, 10000, n_dates),
            'number_of_trades': np.random.normal(20, 5, n_dates),
            'currency_pairs_traded': np.random.randint(3, 15, n_dates),
            'client_spread_revenue': np.random.normal(2000, 500, n_dates),
            'market_volatility': np.random.normal(10, 2, n_dates),
            'client_retention_rate': np.random.normal(0.85, 0.05, n_dates),
            'cross_sell_ratio': np.random.normal(0.3, 0.05, n_dates)
        })

    def train_predictive_models(self):
        """Train predictive models"""
        self.models = {}
        feature_columns = ['spot_volume', 'forward_volume', 'option_volume',
                         'avg_trade_size', 'number_of_trades', 'market_volatility']

        for target in ['fx_volume', 'revenue', 'client_spread_revenue']:
            X = self.data[feature_columns]
            y = self.data[target]

            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

            scaler = StandardScaler()
            X_train_scaled = scaler.fit_transform(X_train)

            model = RandomForestRegressor(n_estimators=100)
            model.fit(X_train_scaled, y_train)

            self.models[target] = {
                'model': model,
                'scaler': scaler,
                'features': feature_columns
            }

    def plot_basic_metrics(self):
        """Plot basic metrics"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('Basic Metrics Dashboard', fontsize=16)

        axes[0, 0].plot(self.data['date'].tail(30), self.data['leads_generated'].tail(30))
        axes[0, 0].set_title('Lead Generation Trend (Last 30 Days)')
        axes[0, 0].set_xticklabels(self.data['date'].tail(30).dt.strftime('%Y-%m-%d'), rotation=45)

        axes[0, 1].plot(self.data['date'].tail(30), self.data['conversion_rate'].tail(30))
        axes[0, 1].set_title('Conversion Rate Trend')
        axes[0, 1].set_xticklabels(self.data['date'].tail(30).dt.strftime('%Y-%m-%d'), rotation=45)

        axes[1, 0].hist(self.data['fx_volume'], bins=30)
        axes[1, 0].set_title('FX Volume Distribution')

        axes[1, 1].scatter(self.data['fx_volume'], self.data['revenue'])
        axes[1, 1].set_title('Revenue vs FX Volume')

        plt.tight_layout()
        plt.show()

    def plot_fx_product_mix(self):
        """Plot FX product mix"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        products = ['Spot', 'Forward', 'Option']
        volumes = [self.data['spot_volume'].mean(),
                  self.data['forward_volume'].mean(),
                  self.data['option_volume'].mean()]

        ax1.bar(products, volumes)
        ax1.set_title('Average Volume by Product')
        ax1.set_ylabel('Volume ($)')

        ax2.stackplot(self.data['date'],
                     self.data['spot_volume'],
                     self.data['forward_volume'],
                     self.data['option_volume'],
                     labels=products)
        ax2.set_title('Product Mix Over Time')
        ax2.legend()

        plt.tight_layout()
        plt.show()

    def plot_predictive_analysis(self):
        """Plot predictive analysis results"""
        fig, axes = plt.subplots(len(self.models), 1, figsize=(15, 5*len(self.models)))

        for i, (target, model_info) in enumerate(self.models.items()):
            X = self.data[model_info['features']].tail(30)
            X_scaled = model_info['scaler'].transform(X)
            predictions = model_info['model'].predict(X_scaled)

            axes[i].plot(self.data['date'].tail(30), self.data[target].tail(30),
                        label='Actual', marker='o')
            axes[i].plot(self.data['date'].tail(30), predictions,
                        label='Predicted', marker='o')

            axes[i].set_title(f'{target} - Actual vs Predicted')
            axes[i].legend()
            axes[i].set_xticklabels(self.data['date'].tail(30).dt.strftime('%Y-%m-%d'),
                                  rotation=45)

        plt.tight_layout()
        plt.show()

    def create_dashboard(self):
        """Create and display the complete dashboard"""
        print("\nFX Sales Analytics Dashboard")
        print("=" * 50)

        print("\nGenerating Basic Metrics...")
        self.plot_basic_metrics()

        print("\nGenerating FX Product Mix Analysis...")
        self.plot_fx_product_mix()

        print("\nGenerating Predictive Analysis...")
        self.plot_predictive_analysis()

dashboard = EnhancedFXSalesDashboard()
dashboard.create_dashboard()
```
